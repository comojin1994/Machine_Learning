```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rc('font', family='AppleGothic')
plt.rc('axes', unicode_minus=False)
```

```python
sns.distplot(train['18~20_ride'], kde=False, bins=50)
plt.axis([0, 50, 0, 450000])
plt.title('퇴근시간 승차 인원 히스토그램')
plt.show()
```

```python
train.groupby('18~20_ride').size().head()
```

```python
train['date'] = pd.to_datetime(train['date'])
train['weekday'] = train['date'].dt.strftime('%a')
```

```python
sns.barplot(x='weekday', y='18~20_ride', data=train)
plt.title('passenger')
plt.show()
```

```python
sns.barplot(x='in_out', y='18~20_ride', data=train)
plt.title('kind')
plt.show()
```

```python
train['8~10_ride'] = train['8~9_ride'] + train['9~10_ride']
eda_data = train.groupby('date')[['18~20_ride', '8~10_ride']].agg('sum').reset_index()
eda_data.head()
```

```python
plt.plot('date', '18~20_ride', 'g-', label='out', data=eda_data)
plt.plot('date', '8~10_ride', 'b-', label='in', data=eda_data)
plt.gcf().auto
```
