```
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import factorial
%matplotlib inline
def poisson(n, lamb):
    pd = (lamb ** n) * math.exp(-lamb) / factorial(n)
    return pd

l = 10
np.random.seed(42)
X1 = np.random.poisson(l, 10000)
np.random.seed(43)
X2 = np.random.poisson(l*2, 10000)
np.random.seed(44)
X3 = np.random.poisson(l, 10000)
plt.figure(figsize=(12, 8))
plt.hist(X1, label='l = 10')
plt.hist(X2, label='l = 20')
plt.hist(X1 + X3, label='X1 + X3')
plt.grid()
plt.legend()
plt.show()
```

```
def f(x):
  if x<0: return 0
  elif 0<=x<2: return x/4
  elif 2<=x<4: return (-x/4) + 1
  else: return 0 
X = np.linspace(-1, 5, 1000)
y = [f(x) for x in X]
plt.plot(X, y, label='f(x)')
plt.axvline(np.median(X), color='red')
plt.axvline(np.sqrt(2), color='green')
plt.axvline(4 - np.sqrt(2), color='green')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```
def F(x):
  if x<0: return 0
  elif 0<=x<2: return (x**2)/8
  elif 2<=x<4: return (-x**2/8) + x - 1
  else: return 1
X = np.linspace(-1, 5, 1000)
y = [F(x) for x in X]
plt.plot(X, y, label='F(x)')
plt.axvline(np.median(X), color='red')
plt.axvline(np.sqrt(2), color='green')
plt.axvline(4 - np.sqrt(2), color='green')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
